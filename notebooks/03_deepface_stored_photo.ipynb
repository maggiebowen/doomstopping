# DeepFace Detecting Emotions

Testing DeepFace on already-taken photos

```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
img_path = "../data/images/happy-maggie.jpg"
img = cv2.imread(img_path)
plt.imshow(img[:, :, ::-1]) # because OpenCV uses BGR (inverted colors)
```

```python
demography = DeepFace.analyze(img_path, actions=["emotion", "age", "gender", "race"])
demography
```

```python
print(f"DeepFace detected {len(demography)} face(s)")
```

```python
face = demography[0]

print(face['dominant_emotion'])
```
